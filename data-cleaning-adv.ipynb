```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder

# Create a DataFrame with a column 'G' for demonstration
df1 = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'G': ['male', 'female', 'female']
})

# Initialize the OneHotEncoder
encoder = OneHotEncoder(sparse=False, drop='first')  # drop='first' to avoid multicollinearity

# Fit and transform the data
one_hot_encoded = encoder.fit_transform(df1[['G']])

# Convert the result to a DataFrame
one_hot_df = pd.DataFrame(one_hot_encoded, columns=encoder.get_feature_names_out(['G']))

# Drop the original 'G' column and concatenate the one-hot encoded DataFrame
df1 = df1.drop(columns=['G'])
df1 = pd.concat([df1, one_hot_df], axis=1)

# Display the updated DataFrame
print(df1)
```

```python
'''Ordinal encoding is a technique used to convert categorical values into numerical values based on their ordinal relationship. This is useful when the categorical variables have a meaningful order or ranking.

In scikit-learn, you can use the OrdinalEncoder class for ordinal encoding. Heres how to apply it:

'''

'''Explanation
Initialize the OrdinalEncoder:

categories=[['low', 'medium', 'high']]: Specifies the order of categories. This is crucial because ordinal encoding assigns integers based on this order.
Fit and Transform:

encoder.fit_transform(df1[['G']]) fits the encoder to the data and transforms the 'G' column into ordinal numerical values.
Add Encoded Column:

The transformed values are added to the DataFrame as a new column 'G_encoded'.
Detailed Breakdown
Creating the DataFrame:

The DataFrame df1 includes a column 'G' with ordinal values.
Initializing the Encoder:

OrdinalEncoder(categories=[['low', 'medium', 'high']]) tells the encoder the specific order of the categories. Each category will be mapped to an integer based on this order (e.g., 'low' -> 0, 'medium' -> 1, 'high' -> 2).
Fitting and Transforming:

fit_transform() both fits the encoder to the data and transforms it into ordinal values.
Adding the Encoded Column:

The encoded values are added to the DataFrame as a new column 'G_encoded'.
This approach is suitable for categorical features where the order matters. If the categories have no meaningful order, one-hot encoding might be more appropriate.'''

import pandas as pd
from sklearn.preprocessing import OrdinalEncoder

# Create a DataFrame with an ordinal column 'G'
df1 = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'G': ['low', 'medium', 'high']
})

# Initialize the OrdinalEncoder
encoder = OrdinalEncoder(categories=[['low', 'medium', 'high']])

# Fit and transform the data
df1['G_encoded'] = encoder.fit_transform(df1[['G']])

# Display the updated DataFrame
print(df1)
```

```python
import pandas as pd
import numpy as np

# Set random seed for reproducibility
np.random.seed(42)

# Create a DataFrame with mixed data types
data = {
    'ID': np.arange(1, 101),  # Integer ID column
    'Age': np.random.randint(18, 70, size=100),  # Random integers for age
    'Salary': np.random.normal(50000, 15000, 100).round(2),  # Normally distributed salary
    'Department': np.random.choice(['HR', 'Engineering', 'Marketing', 'Sales'], 100),  # Categorical department
    'Join_Date': pd.to_datetime(np.random.choice(pd.date_range('2010-01-01', '2020-01-01'), 100)),  # Random dates
    'Gender': np.random.choice(['Male', 'Female'], 100),  # Categorical gender
    'Has_Car': np.random.choice([True, False], 100),  # Boolean feature
    'City': np.random.choice(['New York', 'Los Angeles', 'Chicago', 'Houston'], 100)  # Categorical city
}

df_mixed = pd.DataFrame(data)

# Introduce some missing values
df_mixed.loc[np.random.choice(df_mixed.index, 10), 'Salary'] = np.nan
df_mixed.loc[np.random.choice(df_mixed.index, 5), 'Department'] = np.nan

# Display the first few rows of the dataset
print(df_mixed.head())

# Optionally, save the dataset to a CSV file
df_mixed.to_csv('dataset_mixed_datatypes.csv', index=False)
```
